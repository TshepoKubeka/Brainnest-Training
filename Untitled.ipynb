```python
# Q12: What is the significance value for the above on the Shapiro-Wilk test?
# Answer: To perform a Shapiro-Wilk test on the data, you can use the following code:


stat, p = shapiro(df.loc[df['smoker'] == 1, 'Headcirc'])

print('Shapiro-Wilk test statistic: ', stat)
print('p-value: ', p)

# If the p-value is less than 0.05, then we can reject the null hypothesis that the data is normally distributed. Otherwise, we cannot reject the null hypothesis.

```

```python
#Q13. What is the standard score (Z-score) for head circumference of 35.05 (X=35.05) in non-smoking mothers?
import numpy as np

headcirc = df.loc[df['smoker'] == 0]['Headcirc']
mean = np.mean(headcirc)
std = np.std(headcirc)
X = 35.05
Z = (X - mean) / std
print('Z-score:', Z)
```

```python
#Q15. Are birth weight data for babies of smoking mothers normally distributed?
from scipy.stats import shapiro

birthweight_s = df[df['smoker'] == 1]['Birthweight']
stat, p = shapiro(birthweight_s)
if p > 0.05:
    print('The birth weight data for babies of smoking mothers is normally distributed.')
else:
    print('The birth weight data for babies of smoking mothers is not normally distributed.')
```

```python
#Q17. Based on the dataset you have, how confident can you be in saying that a baby's birth weight will be +/- 1 standard deviation from the mean?

# calculate the mean and standard deviation of birth weight
mean_bw = df['Birthweight'].mean()
std_bw = df['Birthweight'].std()

# calculate the range within 1 standard deviation from the mean
lower = mean_bw - std_bw
upper = mean_bw + std_bw

# calculate the percentage of data within the range
within_range = df[(df['Birthweight'] >= lower) & (df['Birthweight'] <= upper)].shape[0] / df.shape[0] * 100

print('The percentage of birth weight data within +/- 1 standard deviation from the mean is: {:.2f}%'.format(within_range))
```

```python
#Q18. Based on the dataset you have, what is the probability that the birth weight for a baby of a smoking mother will be less than 4.2 kg?
# calculate the proportion of babies of smoking mothers with birth weight less than 4.2 kg
p = df[(df['smoker'] == 1) & (df['Birthweight'] < 4.2)].shape[0] / df[df['smoker'] == 1].shape[0]

print('The probability of a baby of a smoking mother having birth weight less than 4.2 kg is: {:.2f}%'.format(p*100))
```

```python
#Q19. Are data for length of baby of non-smoking mothers normally distributed?
import scipy.stats as stats

# perform Shapiro-Wilk test on length of baby for non-smoking mothers
length_ns = df[df['smoker'] == 0]['Length']
stat, p = stats.shapiro(length_ns)

if p > 0.05:
    print('The data for length of baby of non-smoking mothers is normally distributed with p-value: {:.4f}'.format(p))
else:
    print('The data for length of baby of non-smoking mothers is not normally distributed with p-value: {:.4f}'.format(p))
```

```python
#Q21. What is the standard score for the length of a baby of 48.5cm for non-smoking mothers?
# load the dataset


# calculate the mean and standard deviation of length of baby for non-smoking mothers
mean_len_ns = df[df['smoker'] == 0]['Length'].mean()
std_len_ns = df[df['smoker'] == 0]['Length'].std()

# calculate the Z-score for a length of 48.5cm for non-smoking mothers
z_score = (48.5 - mean_len_ns) / std_len_ns

print("The Z-score for the length of a baby of 48.5cm for non-smoking mothers is:", z_score)

```

```python
#Q22. Based on the dataset you have, what is the probability that the length of baby for non-smoking mothers will be more than 55 cm?
# calculate the probability of a length greater than 55cm for non-smoking mothers
prob_greater_than_55cm = 1 - stats.norm.cdf((55 - mean_len_ns) / std_len_ns)

print("The probability of a length greater than 55cm for non-smoking mothers is:", prob_greater_than_55cm)
```
